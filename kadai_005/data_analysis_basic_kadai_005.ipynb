```python
#matplotlib.pyplotメソッドとnumpyをインポートするコードを記述する
import matplotlib.pyplot as plt
import numpy as np

#複数グラフを表示
fig, axes = plt.subplots(nrows=2, ncols=3,tight_layout = True)

#bar
y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)

#sin
x = np.linspace(0, 10, 100)
y = 2 + 2 * np.sin(2 * x)
axes[0,1].set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))
axes[0,1].plot(x, y)

#hist
x02 = np.random .normal(15, 5, 2000)
axes[0,2].hist(x02, bins=30)
axes[0, 2].set_xlim(-10, 40)

#scatter
np.random.seed(3)
x3 = 4 + np.random.normal(0, 2, 60)
y3 = 4 + np.random.normal(0, 2, len(x3))
axes[1,0].scatter(x3, y3)

plt.show()
```
